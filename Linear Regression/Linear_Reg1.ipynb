```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Data
hours_of_study = np.array([2,5,7,9,12]).reshape(-1,1)
hours_of_study
```

```python
exam_scores = np.array([60,65,70,75,80])
exam_scores
```

```python
# Create a linear regression model
model = LinearRegression()
```

```python
# Fit the model into data
model.fit(hours_of_study,exam_scores)
```

```python
# Get the slope m and y-intercept b
slope = model.coef_[0]
intercept = model.intercept_
```

```python
# Make predictions
hours_to_predict = np.array([8]).reshape(-1,1)
predicted_exam_score=model.predict(hours_to_predict)
predicted_exam_score
```

```python
# Plot the data
plt.scatter(hours_of_study,exam_scores,color='blue',label='Actual Data')
plt.plot(hours_of_study,model.predict(hours_of_study),color='red',label='Regression Line')
plt.scatter(hours_to_predict,predicted_exam_score,color='green',marker='x',label='Prediction')
# Add labels
plt.xlabel("Hours of Study")
plt.ylabel("Exam Score")
plt.show()
```
